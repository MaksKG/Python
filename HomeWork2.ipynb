## –ó–∞–≤–¥–∞–Ω–Ω—è
–†–µ–∞–ª—ñ–∑—É–π—Ç–µ –≤–µ—Å—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–π –≤–∏—â–µ —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª –≤ –ø–µ–≤–Ω–æ–º—É –≤–∏–¥—ñ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ñ—É–Ω–∫—Ü—ñ—ó, –∫–ª–∞—Å–∏ —Ç–∞ –û–û–ü.

```python
class Health:
    __health = 100
    __max_health = 100

    def get_health(self):
        return self.__health
    
    def set_health(self, health):
        self.__health = health

    def __init__(self,  max_health=100):
        self.__max_health = max_health
        if(self.__health > self.__max_health):
            self.__health = self.__max_health

    def damage(self, damage):
        self.__health -= damage
        if(self.__health < 0):
            self.__health = 0
    
    def heal(self, heal):
        self.__health += heal
        if(self.__health > self.__max_health):
            self.__health = self.__max_health


class Move:
    __position = [0, 0]
    __move_speed = 1
    def get_position(self):
        return self.__position
    
    def move(self, x, y):
        if(x > 1):
            x = 1
        elif(x < -1):
            x = -1
        if(y > 1):
            y = 1
        elif(y < -1):
            y = -1

        self.__position[0] += x * self.__move_speed
        self.__position[1] += y * self.__move_speed

    def __init__(self, x=0, y=0, move_speed=1):
        self.__position = [x, y]
        self.__move_speed = move_speed

class Fighter: 
    __damage = 20

    def get_damage(self):
        return self.__damage
    def attack(self, target:Health):
        target.damage(self.__damage)

    def __init__(self, damage=10):
        self.__damage = damage
#)____________________________________________________________
class Bomber:
    __damage = 50

    def get_damage(self):
        return self.__damage
    def attack(self, target:Health):
        target.damage(self.__damage)

    def __init__(self, damage=50):
        self.__damage = damage
#)____________________________________________________________

# class Base:
#     __position = [0, 0]
#     __move_speed = 0
#     def get_position(self):
#         return self.__position
    
#     def __init__(self, x=0, y=0):
#         self.__position = [x, y]
#)____________________________________________________________        

import random
class Name_F:
    __name = "default"
    __id = random.randint(0, 100)
    def get_name(self):
        return self.__name
    def __init__(self, _id = random.randint(0, 100), name=f'fighter'):
        if(name == 'fighter'):
            name = f'fighter [{_id}]'
        self.__name = name

import random
class Name_B:
    __name = "default"
    __id = random.randint(0, 100)
    def get_name(self):
        return self.__name
    def __init__(self, _id = random.randint(0, 100), name=f'bomber'):
        if(name == 'bomber'):
            name = f'bomber [{_id}]'
        self.__name = name

import random
class Name_Base:
    __name = "default"
    __id = random.randint(0, 100)
    def get_name(self):
        return self.__name
    def __init__(self, _id = random.randint(0, 100), name=f'Base'):
        if(name == 'Base'):
            name = f'Base [{_id}]'
        self.__name = name
        
def generate_plane():
    import random

    random_value = random.randint(0, 100)

    if(random_value < 50):
        class Plane(Health, Move, Name_B,Bomber):
            def move(self, x, y):
                if(self.get_health() > 0):
                    return super().move(x, y)
            def __init__(self, max_health=100, x=0, y=0, move_speed=1, damage =50):
                Health.__init__(self, max_health)
                Move.__init__(self, x, y, move_speed) 
                Bomber.__init__(self, damage)
                Name_B.__init__(self, random.randint(0, 100))

        return Plane
    else:
        class Plane(Health, Move, Fighter, Name_F):
            def move(self, x, y):
                if(self.get_health() > 0):
                    return super().move(x, y)
            def __init__(self, max_health=100, x=0, y=0, move_speed=1, damage=20):
                Health.__init__(self, max_health)
                Move.__init__(self, x, y, move_speed) 
                Fighter.__init__(self, damage)
                Name_F.__init__(self, random.randint(0, 100))

        return Plane
    

def generate_base():
    import random

    random_value = random.randint(0, 100)

    if(random_value < 50):
        class Base(Health,Move,Name_Base):
            def move(self, x, y):
                if(self.get_health() > 0):
                    return super().move(x, y)
            def __init__(self, max_health=100, x=0, y=0, move_speed=0):
                Health.__init__(self, max_health)
                Move.__init__(self, x, y, move_speed) 
                Name_Base.__init__(self, random.randint(0, 100))
        return Base
    else:
        class Base(Health,Move,Name_Base):
            def move(self, x, y):
                if(self.get_health() > 0):
                    return super().move(x, y)
            def __init__(self, max_health=200, x=0, y=0, move_speed=0):
                Health.__init__(self, max_health)
                Move.__init__(self, x, y, move_speed) 
                Name_Base.__init__(self, random.randint(0, 100))
        return Base
```
